# Principal Component Analysis

1. Import required modules

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn import datasets
import timeit
```

2. Load dataset

```python
iris = datasets.load_iris()
```

3. Split data and target

```python
# Separating out the features
X = iris.data[:, :4]
# Separating out the target
y = iris.target
```

4. Change the quantity of knowledge that you want to keep it

```python
information = [0.4, 0.99]
```

5. By iterate over each parameter you can now implement PCA and differentiate between number of features for each knowledge

```python
for i in information:
    start = timeit.default_timer()

    # Splitting the data into training and testing
    X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=1/7.0, random_state=0)
    
    # Standardizing the features
    scaler = StandardScaler()
    scaler.fit(X_train)
    X_train = scaler.transform(X_train)
    X_test = scaler.transform(X_test)

    # Applying PCA
    pca = PCA(i)
    pca.fit(X_train)

    X_train = pca.transform(X_train)
    print("PCA components:",pca.n_components_) 
    X_test = pca.transform(X_test)

    # Fitting Logistic Regression to the Training set
    logisticRegr = LogisticRegression()
    logisticRegr.fit(X_train, y_train)
    print("score:",logisticRegr.score(X_test, y_test))
    
    stop = timeit.default_timer()
    print('Time: ', stop -start )

    print("")

```
